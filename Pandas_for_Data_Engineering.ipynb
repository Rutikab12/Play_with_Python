#Merge DataFrames
same meaning as sQL joins.

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35, 38],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "humidity": [65,68,71],
})
df2
```

```python
#working as inner join basically
df=pd.merge(df1,df2,on='city')
df
```

```python
#outer join
df=pd.merge(df1,df2,on='city',how='outer')
df
```

```python
#left join
df=pd.merge(df1,df2,on='city',how='left')
df
```

```python
#right join
df=pd.merge(df1,df2,on='city',how='right')
df
```

```python
#how will help in chosing right option for merging both dataframes
#with indicator=True, you will get to know from which dataframe each row in df coming from.
df=pd.merge(df1,df2,on='city',how='outer',indicator=True)
df
```

```python
#using suffixes same as aliases in SQL
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35,38],
    "humidity": [65,68,71, 75]
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "temperature": [21,14,35],
    "humidity": [65,68,71]
})
df2
```

```python
df3= pd.merge(df1,df2,on="city",how="outer", suffixes=('_first','_second'))
df3
```

```python
#join
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
})
df1.set_index('city',inplace=True)
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","orlando"],
    "humidity": [65,68,75],
})
df2.set_index('city',inplace=True)
df2
```

```python
df1.join(df2,lsuffix='_l', rsuffix='_r')
```

```python
import sqlalchemy
engine = sqlalchemy.create_engine('mysql+pymysql://root:@localhost:3306/application')
```

image.png

```python
#read whole table
df=pd.read_sql_table('customers',engine)
```

```python
#read only selected columns
df = pd.read_sql_table('customers', engine, columns=["name"])
df
```

```python
#joining two table and read them in dataframe
query = '''
 SELECT customers.name, customers.phone_number, orders.name, orders.amount
 FROM customers INNER JOIN orders
 ON customers.id=orders.customer_id
'''
df = pd.read_sql_query(query,engine)
df
#read_sql is a wrapper around read_sql_query and read_sql_table
```

```python
#Write to mysql database using to_sql
df=pd.read_csv('/content/sample_data/customers.csv')
df
```
