#Reading/Writing CSV , Excel files using Pandas


```python
#reading csv file
import pandas as pd
#df=pd.read_csv('/content/sample_data/stock_data.csv',nrows=3) #for reading only 3 rows
#handling missing or messy data while reading
#df=pd.read_csv('/content/sample_data/stock_data.csv',na_values=['not available','n.a.'])
#providing dict to fillna column wise
df=pd.read_csv('/content/sample_data/stock_data.csv',na_values={
    'eps':['not available','n.a.'],
    'revenue':['not available','n.a.',-1],
    'people':['not available','n.a.'],
    'price':['not available','n.a.']
})
#saving into csv files
df.to_csv('new.csv',index=False)
```

```python
#reading and writing excel file
df=pd.read_excel('/content/sample_data/stock_data.xlsx','Sheet1')
df
```

```python
#using a converter for filling up na values in people cell with meaningful data
def convert_people_cell(cell):
  if cell=='n.a.':
    return 'sam walton'
  return cell

df=pd.read_excel('/content/sample_data/stock_data.xlsx','Sheet1',converters={
    'people': convert_people_cell
})
df.to_excel('new.xlsx',sheet_name='Stocks',index=False)
```

```python
#writing data of two dataframes into separate sheets in excel file
df_stocks = pd.DataFrame({
    'tickers': ['GOOGL', 'WMT', 'MSFT'],
    'price': [845, 65, 64 ],
    'pe': [30.37, 14.26, 30.97],
    'eps': [27.82, 4.61, 2.12]
})

df_weather =  pd.DataFrame({
    'day': ['1/1/2017','1/2/2017','1/3/2017'],
    'temperature': [32,35,28],
    'event': ['Rain', 'Sunny', 'Snow']
})
```

```python
with pd.ExcelWriter('stocks_weather.xlsx') as writer:
    df_stocks.to_excel(writer, sheet_name="stocks",index=False)
    df_weather.to_excel(writer, sheet_name="weather",index=False)
```

#Handling Missing Data in Dataframe

```python
df=pd.read_csv('/content/sample_data/weather_data (1).csv',parse_dates=['day'])
df
```

```python
#filling null values with different methods
df.fillna(0)
```

```python
df.fillna({
    'temperature':0,
    'windspeed':0,
    'event':'no event'
})
```

```python
#using forward fill & backward fill for 0.0, with axis=columns it will copy values for NaN from adjacent column
df.fillna(method='ffill')

df.fillna(method='bfill',axis='columns')
```

```python
#using linear interpolate for better gusess to fill NaN values
df.interpolate()
#new_df = df.interpolate(method="time")
df
```

```python
#dropna, droping NaN fiels
df.dropna(how='all') #don;t drop columns ehich have values for that row

df.dropna(thresh=1) #meaning row with NaN should have at least one value to be kept otherwise drop it
```

```python
#filling daterange
df=pd.date_range('01-01-2017','01-11-2017')
idx=pd.DatetimeIndex(df)
df=df.reindex(idx)
df
```
